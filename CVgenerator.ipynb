In this second cell all the text items to include in the CV are grouped by category in lists of lists to later on automatize the formatting of the CV.

Define style and generates the CV

```python
import matplotlib.pyplot as plt
%matplotlib inline

# set font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'STIXGeneral'
color='teal'    #this is the main color of the cv, it can be change for the whole document from here

fig, ax = plt.subplots(figsize=(8.5, 12))

# Background boxes and sections
from matplotlib.patches import Rectangle
ax.add_patch(Rectangle((.0, .0), 0.3, 1,   linewidth=1, edgecolor=color, facecolor=color, alpha=0.1))
ax.add_patch(Rectangle((.15, .89), 0.7, 0.09, linewidth=1, color='w'))
ax.add_patch(Rectangle((.15, .89), 0.7, 0.09, linewidth=1, facecolor='none',edgecolor='k'))

# set background color
ax.set_facecolor('white')

# remove axes
plt.axis('off')



# ADD TEXT
plt.annotate(Annotation                     , (0.98,.99)  , weight='regular' , fontsize=7  , ha='right'  , alpha=0.75)
plt.annotate(Name                           , (.5,.94)    , weight='bold'    , fontsize=20 , ha='center')
plt.annotate(Title                          , (.5,.91)    , weight='regular' , fontsize=14 , ha='center')

#LEFT COLUMN
y=.84  #initialize the left column y value
for section in left:
    plt.annotate(section[0]                , (.03,y)   , weight='bold'    , fontsize=10 , color=color)
    ax.axhline(y=y-0.008, xmin=.03, xmax=.27, color=color)
    y=y-0.03
    if section[0]=="PROFESSIONAL ID":
        for item,url in zip(section[1:],urls):
            plt.annotate(item              , (.03,y)   , weight='demibold', fontsize=8  , alpha= 0.5 , color=color , url=url)
            y=y-0.02
    else:
        for subsection in section[1:]:
            if isinstance(subsection,list):
                plt.annotate(subsection[0] , (.03,y)   , weight='bold'    , fontsize=8)
                y=y-0.02
                for item in subsection[1:]:
                    plt.annotate(item      , (.03,y)   , weight='regular' , fontsize=8)
                    y=y-0.02
                y=y-0.01
            else:
                plt.annotate(subsection    , (.03,y)   , weight='regular' , fontsize=8)
                y=y-0.02
    y=y-0.03


#PROFILE
ax.add_patch(Rectangle((.32, .832), 0.66, 0.025, color=color, alpha=1))
plt.annotate(ProfileHeader    , (.33,.84)   , weight='bold'    , fontsize=10 , color='w')
plt.annotate(Profile          , (.33,.818)  , weight='regular' , fontsize=9  , va='top')


#MAIN
y=0.583     #this value has to be updated according to the lenght of the profile description
for section in main:
    plt.annotate(section[0]                , (.33,y)   , weight='bold'    , fontsize=10 , color='w')
    ax.add_patch(Rectangle((.32, y-0.008), 0.66, 0.025, color=color, alpha=1))
    y=y-0.03
    for subsection in section[1:]:
        plt.annotate(subsection[0]         , (.33,y)   , weight='bold'    , fontsize=10)
        if len(subsection[1])>0:
            y=y-0.015
            plt.annotate(subsection[1]         , (.33,y)   , weight='regular' , fontsize=9  , alpha=.7)
        y=y-0.015
        if len(subsection[2])>0:
            for item in subsection[2]:
                plt.annotate(item              , (.34,y)   , weight='regular' , fontsize=9)
                y=y-0.015
        y=y-0.02    
    y=y-0.01


plt.savefig('resumeexample.pdf', dpi=300,bbox_inches='tight')
```
